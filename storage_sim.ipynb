```python
def gen_polygon_mask(lats, lons, res, geom):
    gridcells = [Polygon(((x-res, y-res), 
                      (x-res, y+res), 
                      (x+res, y+res), 
                      (x+res,y-res), 
                      (x-res, y-res))) for x,y in zip(lons, lats)]
    ngrd = len(gridcells)
    mask = np.zeros((ngrd,), dtype=int)
    for ii in range(ngrd):
        mask[ii] = geom.intersects(gridcells[ii])
    print(mask.sum(), 'intersections')
    return mask
```

```python
def calc_grace_basin_mean(bounds, geom):
    grace_jpl_bbox = grace_jpl.sel(lat=slice(bounds['miny'], bounds['maxy']), 
                                   lon=slice(bounds['minx']+360, bounds['maxx']+360))
    grace_csr_bbox = grace_csr.sel(lat=slice(bounds['miny'], bounds['maxy']), 
                                   lon=slice(bounds['minx']+360, bounds['maxx']+360))
    grace_gsfc_bbox = grace_gsfc.sel(lat=slice(bounds['miny'], bounds['maxy']), 
                                   lon=slice(bounds['minx']+360, bounds['maxx']+360))
    
    lons, lats = np.meshgrid(grace_jpl_bbox.lon, grace_jpl_bbox.lat)
    lons[lons>=180] = lons[lons>=180]-360
    mask = gen_polygon_mask(lats.ravel(), lons.ravel(), 0.5, geom)
    mask_jpl = xr.DataArray(mask.reshape(lats.shape), dims=(('lat','lon')), name='mask')

    lons, lats = np.meshgrid(grace_csr_bbox.lon, grace_csr_bbox.lat)
    lons[lons>=180] = lons[lons>=180]-360
    mask = gen_polygon_mask(lats.ravel(), lons.ravel(), 0.25, geom)
    mask_csr = xr.DataArray(mask.reshape(lats.shape), dims=(('lat','lon')), name='mask')
    
    lons, lats = np.meshgrid(grace_gsfc_bbox.lon, grace_gsfc_bbox.lat)
    lons[lons>=180] = lons[lons>=180]-360
    mask = gen_polygon_mask(lats.ravel(), lons.ravel(), 0.5, geom)
    mask_gsfc = xr.DataArray(mask.reshape(lats.shape), dims=(('lat','lon')), name='mask')

    lwe_jpl = grace_jpl_bbox['lwe_thickness'].where(mask_jpl>0).mean(('lon','lat'))*10
    lwe_csr = grace_csr_bbox['lwe_thickness'].where(mask_csr>0).mean(('lon','lat'))*10
    lwe_gsfc = grace_gsfc_bbox['lwe_thickness'].where(mask_gsfc>0).mean(('lon','lat'))*10
    
    return lwe_jpl, lwe_csr, lwe_gsfc
```

```python
dict_plot = {}

# CR
tws_mdl = h0_opt_cr['TWS']
lons, lats = np.meshgrid(tws_mdl.lon, tws_mdl.lat)
for case in ('cr','lcrb','ucrb'):
    print(case)
    if case=='lcrb':
        jpl_cr, csr_cr, gsfc_cr = calc_grace_basin_mean(bounds_lcrb, geom_lcrb)
        mask_mdl = xr.open_dataarray('./mask_lcrb.nc')
    if case=='ucrb':
        jpl_cr, csr_cr, gsfc_cr = calc_grace_basin_mean(bounds_ucrb, geom_ucrb)
        mask_mdl = xr.open_dataarray('./mask_ucrb.nc')
    if case=='cr':
        jpl_cr, csr_cr, gsfc_cr = calc_grace_basin_mean(bounds_cr, geom_cr)
        mask_mdl = xr.DataArray(np.ones(lats.shape), dims=(('lat','lon')), name='mask')

    tdelta_base = slice('2004-01', '2009-12')
    tws_mdl = h0_opt_cr['TWS']
    lons, lats = np.meshgrid(tws_mdl.lon, tws_mdl.lat)
    tws_mdl_ma = tws_mdl.where(mask_mdl>0).mean(('lon','lat')).compute()
    tws_mdl_anom_opt = tws_mdl_ma - tws_mdl_ma.sel(time=tdelta_base).mean()
    tws_mdl_anom_opt['time'] = tws_mdl_anom_opt.time.astype('datetime64[ns]')
    
    tws_mdl = h0_base_cr['TWS']
    tws_mdl_ma = tws_mdl.where(mask_mdl>0).mean(('lon','lat')).compute()
    tws_mdl_anom_base = tws_mdl_ma - tws_mdl_ma.sel(time=tdelta_base).mean()
    tws_mdl_anom_base['time'] = tws_mdl_anom_base.time.astype('datetime64[ns]')

    jpl_cr['time'] = jpl_cr.time.astype('datetime64[D]')
    csr_cr['time'] = csr_cr.time.astype('datetime64[D]')
    gsfc_cr['time'] = gsfc_cr.time.astype('datetime64[D]')

    ## Missing dates
    tdelta_eval = slice('2002-04', '2020-09')
    yy = tws_mdl_anom_opt.sel(time=tdelta_eval)
    yy_base = tws_mdl_anom_base.sel(time=tdelta_eval)
    da_jpl = jpl_cr.resample(time='1M').mean().sel(time=tdelta_eval)
    da_csr = csr_cr.resample(time='1M').mean().sel(time=tdelta_eval)
    da_gsfc = gsfc_cr.resample(time='1M').mean().sel(time=tdelta_eval)
    da_jpl['time'] = yy.time
    da_csr['time'] = yy.time
    da_gsfc['time'] = yy.time
    da_grace = xr.concat([da_jpl,da_csr,da_gsfc], dim='inst')
    time = da_grace.time.astype('datetime64[m]').data
    dict_plot[case] = (yy_base,yy,da_jpl,da_csr,da_gsfc,time,da_grace)
```

```python
# CA
mask_mdl = xr.open_dataarray('./mask_ca.nc')
tdelta_base = slice('2004-01', '2009-12')
tws_mdl = h0_opt_ca['TWS']
tws_mdl_ma = tws_mdl.mean(('lon','lat')).compute()
tws_mdl_anom_opt = tws_mdl_ma - tws_mdl_ma.sel(time=tdelta_base).mean()
tws_mdl_anom_opt['time'] = tws_mdl_anom_opt.time.astype('datetime64[ns]')

tws_mdl = h0_base_ca['TWS']
tws_mdl_ma = tws_mdl.where(mask_mdl>0).mean(('lon','lat')).compute()
tws_mdl_anom_base = tws_mdl_ma - tws_mdl_ma.sel(time=tdelta_base).mean()
tws_mdl_anom_base['time'] = tws_mdl_anom_base.time.astype('datetime64[ns]')

jpl_ca['time'] = jpl_ca.time.astype('datetime64[D]')
csr_ca['time'] = csr_ca.time.astype('datetime64[D]')
gsfc_ca['time'] = gsfc_ca.time.astype('datetime64[D]')

## Missing dates
tdelta_eval = slice('2002-04', '2020-09')
yy = tws_mdl_anom_opt.sel(time=tdelta_eval)
yy_base = tws_mdl_anom_base.sel(time=tdelta_eval)
da_jpl = jpl_ca.resample(time='1M').mean().sel(time=tdelta_eval)
da_csr = csr_ca.resample(time='1M').mean().sel(time=tdelta_eval)
da_gsfc = gsfc_ca.resample(time='1M').mean().sel(time=tdelta_eval)
da_jpl['time'] = yy.time
da_csr['time'] = yy.time
da_gsfc['time'] = yy.time

da_grace = xr.concat([da_jpl,da_csr,da_gsfc], dim='inst')
time = da_grace.time.astype('datetime64[m]').data
dict_plot['ca'] = (yy_base,yy,da_jpl,da_csr,da_gsfc,time,da_grace)
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,4), dpi=100, sharex=True)
plt.subplots_adjust(hspace=.25)
for i, case in enumerate(('cr','ca','lcrb','ucrb')):
    ax = axes.ravel()[i]
    yy_base,yy,da_jpl,da_csr,da_gsfc,time,da_grace = dict_plot[case]
    ax.axhline(0, c='k', lw=1, ls='--')
    yy_base.plot(ax=ax,lw=1, label='Reference')
    g = yy.plot(ax=ax,lw=1, c='tab:red', label='Optimized')
    da_jpl.plot(ax=ax,lw=1, c='k', label='GRACE JPL')
    # da_csr.plot(c='grey', label='CSR')
    # da_gsfc.plot(c='lightgrey', label='GSFC')
    ax.fill_between(time, da_grace.min('inst'), da_grace.max('inst'), 
                     color='grey',edgecolor='k', lw=.5,alpha=.5, label='GRACE range')
    ax.set_ylabel('(mm)')
    ax.set_xlabel('')
    ax.tick_params(axis='x', rotation=0)
    # ax.set_xticklabels(, rotation=0, ha='center')
    ax.margins(x=0, y=0.05)
    
    if case=='cr':
        ax.set_title('Colorado Basin', loc='left')
    elif case=='lcrb':
        ax.set_title('Lower Colorado Basin', loc='left')
    elif case=='ucrb':
        ax.set_title('Upper Colorado Basin', loc='left')
    else:
        ax.set_title('California Basin', loc='left')
h,l = ax.get_legend_handles_labels()
fig.legend(h,l, loc='upper center', ncol=4, bbox_to_anchor=(.5,1.1))
```

```python
def trend_line(da):
    yhat = da.dropna('time')
    coef = np.polyfit(np.arange(yhat.size),yhat,1)
    poly1d_fn = np.poly1d(coef)
    return poly1d_fn(np.arange(da.size))
```

```python
fig, axes_all = plt.subplots(2,4, figsize=(13,4), dpi=100, sharex=False, layout="constrained",
                         gridspec_kw={'width_ratios':[1,.25,1,.25]},
                         )
# plt.subplots_adjust(hspace=.5)
axes_ts = axes_all[:, ::2]
axes_ac = axes_all[:, 1::2]

months = np.arange(12)
idx = np.append(months[9:],months[:9])

for i, case in enumerate(('cr','ca','lcrb','ucrb')):
    ax = axes_ts.ravel()[i]
    yy_base,yy,da_jpl,da_csr,da_gsfc,time,da_grace = dict_plot[case]
    # ax.axhline(0, c='k', lw=1, ls='--')
    yy_base.plot(ax=ax,lw=1, label='Reference')
    g = yy.plot(ax=ax,lw=1, c='tab:red', label='Optimized')
    da_jpl.plot(ax=ax,lw=1, c='k', label='GRACE JPL')
    # da_csr.plot(c='grey', label='CSR')
    # da_gsfc.plot(c='lightgrey', label='GSFC')
    ax.fill_between(time, da_grace.min('inst'), da_grace.max('inst'), 
                     color='grey',edgecolor='k', lw=.5,alpha=.5, label='GRACE range')
    
    ax.plot(da_jpl.time, trend_line(da_jpl), '--k', dashes=(20, 5), lw=1)
    ax.plot(yy_base.time, trend_line(yy_base), c='tab:blue', ls='--', dashes=(20, 5), lw=1)
    ax.plot(yy.time, trend_line(yy), c='tab:red', ls='--', dashes=(20, 5), lw=1)
    
    ax.set_ylabel('(mm)')
    ax.set_xlabel('')
    ax.tick_params(axis='x', rotation=0)
    # ax.set_xticklabels(, rotation=0, ha='center')
    ax.margins(x=0, y=0.05)
    
    ax_ac = axes_ac.ravel()[i]
    yy_base_ac = yy_base.groupby('time.month').mean()
    yy_ac = yy.groupby('time.month').mean()
    da_jpl_ac = da_jpl.groupby('time.month').mean()
    da_csr_ac = da_csr.groupby('time.month').mean()
    da_gsfc_ac = da_gsfc.groupby('time.month').mean()
    da_grace_ac = da_grace.groupby('time.month').mean()
    
    ax_ac.plot(yy_base_ac.data[idx], lw=1, label='Reference')
    ax_ac.plot(yy_ac.data[idx], c='tab:red', lw=1, label='Optimized')
    ax_ac.plot(da_jpl_ac.data[idx], c='k', lw=1, label='GRACE JPL')
    ax_ac.fill_between(months, da_grace_ac.min('inst').data[idx], da_grace_ac.max('inst').data[idx], 
                         color='grey',edgecolor='k', lw=.5, alpha=.5, label='GRACE range')
    ax_ac.set_xticks(months)
    ax_ac.set_xticklabels(['O','N','D', 'J','F','M','A','M','J','J','A','S',]);
    ax_ac.margins(x=0, y=0.05)
    
    if case=='cr':
        ax.set_title('a) Colorado Basin', loc='left')
    elif case=='lcrb':
        ax.set_title('c) Lower Colorado Basin', loc='left')
    elif case=='ucrb':
        ax.set_title('d) Upper Colorado Basin', loc='left')
    else:
        ax.set_title('b) California Basin', loc='left')
h,l = ax.get_legend_handles_labels()
fig.legend(h,l, loc='upper center', ncol=4, bbox_to_anchor=(.5,1.1), frameon=False)
```
